```python
import os
import numpy as np
import cv2

# Define paths
train_dir = "D:/Major Project/normalized/train"
valid_dir = "D:/Major Project/normalized/valid"

# Function to convert .npy files to images
def convert_npy_to_images(directory):
    for file in os.listdir(directory):
        if file.endswith(".npy"):
            file_path = os.path.join(directory, file)
            image_array = np.load(file_path)  # Load NumPy array
            
            # Convert file name from .npy to .jpg
            image_name = file.replace(".npy", ".jpg")
            image_path = os.path.join(directory, image_name)

            # Save the image
            cv2.imwrite(image_path, image_array)

            # Remove the .npy file
            os.remove(file_path)

            print(f"Converted {file} to {image_name}")

# Convert images in train and valid directories
convert_npy_to_images(train_dir)
convert_npy_to_images(valid_dir)
```
