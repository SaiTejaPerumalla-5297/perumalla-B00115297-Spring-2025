```python
import os
import pandas as pd
import shutil

# Path to dataset
base_dir = r"D:\Major Project\normalized\train"
labels_file = os.path.join(base_dir, "labels.csv")

# Read labels.csv
df = pd.read_csv(labels_file)

# Ensure class folders exist
for category in df["class"].unique():
    class_folder = os.path.join(base_dir, category)
    os.makedirs(class_folder, exist_ok=True)  # Create folder if it doesn’t exist

# Move images to correct class folders
for _, row in df.iterrows():
    img_name = row["filename"]
    img_class = row["class"]
    src_path = os.path.join(base_dir, img_name)  # Current location
    dest_path = os.path.join(base_dir, img_class, img_name)  # Target location

    if os.path.exists(src_path):  # Move only if file exists
        shutil.move(src_path, dest_path)
    else:
        print(f"⚠ Warning: {src_path} not found.")

print("✅ Images successfully moved to their respective class folders.")
```
